# Setup

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

# Confusion Matrix

```python
labels = np.load("../data/iemocap/labels.npy", allow_pickle = True).tolist()
labels.pop(7)
cm = np.load('../results/co_learning_bilstm_im/cm_res.npy')
cm_bieflstm_0 = cm[0]
cm_bieflstm_80 =cm[4]
```

```python
#None Dropout
df_cm = pd.DataFrame(cm_bieflstm_0, index = [i for i in labels ],
                      columns = [i for i in labels])
plt.title("0% Dropout")
sn.heatmap(df_cm, annot=True, fmt= 'd', cmap = 'Blues')
plt.savefig('../results/co_learning_bilstm_im/cm_res_0')
```

```python
#80% dropout
df_cm = pd.DataFrame(cm_bieflstm_80, index = [i for i in labels ],
                      columns = [i for i in labels])
sn.heatmap(df_cm, annot=True, fmt= 'd', cmap = 'Blues')
plt.title("80% Dropout")
plt.savefig('../results/co_learning_bilstm_im/cm_res_80')
```
